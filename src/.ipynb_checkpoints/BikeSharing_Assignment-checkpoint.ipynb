```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import sklearn
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.metrics import r2_score

import warnings
warnings.filterwarnings('ignore')

# Importing RFE and LinearRegression
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
data_path = r"E:\ml_nlp\git\DummyFolder\assets\day.csv"
bike_share = pd.read_csv(data_path)
bike_share.head()
```

```python
bike_share.shape
```

```python
bike_share.info()
```

```python
bike_share.describe()
```

```python
# Visualising numeric variable using pairplot
num_var = bike_share[['temp','atemp','hum','windspeed','cnt']]
sns.pairplot(num_var)
plt.show()
```

```python
# Visualising categorical data using boxplot
plt.figure(figsize = (20,12))

plt.subplot(2,3,1)
sns.boxplot(x='workingday', y='cnt', data=bike_share)
plt.subplot(2,3,2)
sns.boxplot(x='weekday', y='cnt', data=bike_share)
plt.subplot(2,3,3)
sns.boxplot(x='holiday', y='cnt', data=bike_share)
plt.subplot(2,3,4)
sns.boxplot(x='mnth', y='cnt', data=bike_share)
plt.subplot(2,3,5)
sns.boxplot(x='season', y='cnt', data=bike_share)
plt.subplot(2,3,6)
sns.boxplot(x='weathersit', y='cnt', data=bike_share)
plt.show()
```

```python
# Removing unwanted columns
new_bike_share = bike_share.drop(['instant','dteday','casual','registered'], axis=1)
new_bike_share.info()
```

```python
# Categorising data

new_bike_share['season'] = new_bike_share['season'].astype('category')
new_bike_share['weathersit'] = new_bike_share['weathersit'].astype('category')
new_bike_share['mnth'] = new_bike_share['mnth'].astype('category')
new_bike_share['weekday'] = new_bike_share['weekday'].astype('category')

new_bike_share.info()
```

```python
# Create dummy variable for categorical data in dataset

new_bike_share = pd.get_dummies(new_bike_share, drop_first=True)

new_bike_share.info()
```

```python
# Correlation Matrix

plt.figure(figsize = (25,20))
sns.heatmap(new_bike_share.corr(),annot=True,cmap="YlGnBu")
```

```python
df_train, df_test = train_test_split(new_bike_share, train_size = 0.7, random_state=100)
print(df_train.shape)
print(df_test.shape)
```

```python
# Instantiating an Object
scaler = MinMaxScaler()

df_train.head()
```

```python
# Create list of numeric variables

num_vars = ['temp','atemp','hum','windspeed','cnt']

# Fit on data
df_train[num_vars] = scaler.fit_transform(df_train[num_vars])

df_train.head()
```

```python
df_train.describe()
```

```python
# Dividing X and y sets to build model

y_train = df_train.pop('cnt')
X_train = df_train
```

```python
# Performing RFE to model

lm = LinearRegression()
lm.fit(X_train, y_train)
rfe = RFE(lm,15)
rfe = rfe.fit(X_train, y_train)
```

```python
list(zip(X_train.columns, rfe.support_, rfe.ranking_))
```

```python
col = X_train.columns[rfe.support_]
col
```

```python
X_train.columns[~rfe.support_]
```

```python
# Creating X_test dataframe with rfe selected variables

X_train_rfe = X_train[col]
```

### Model 1

```python
# Adding a constant
X_train_rfe = sm.add_constant(X_train_rfe)
```

```python
# Fitting the model
lr_model1 = sm.OLS(y_train, X_train_rfe).fit()
```

```python
lr_model1.params
```

```python
print(lr_model1.summary())
```

```python
# Calculating the VIFs for the model

vif = pd.DataFrame()
X = X_train_rfe
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values,i) for i in range (X.shape[1])]
vif['VIF'] = round(vif['VIF'],2)
vif = vif.sort_values(by="VIF", ascending=False)
vif
```
